# **TensoRt for video1 and 2 for yolov5n**

```python
import torch
import cv2
import random
import time
import pathlib
from ultralytics import YOLO
from ultralytics.utils import LOGGER
from ultralytics.nn.autobackend import AutoBackend
import matplotlib.pyplot as plt

# Define the TensorRT detection function
def tensorrt_detection(model, source, image):
    im = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Preprocess the input image (example)
    preds = model(im)  # Perform inference
    results = preds[0]  # Assuming the output is the same as YOLO
    d = results.boxes  # Extract detected boxes

    # Handle multiple or single detections
    tensor_size = d.cls.size()[0]
    if tensor_size > 1:
        cls, conf, box = d.cls.squeeze(), d.conf.squeeze(), d.xyxy.squeeze()
    else:
        cls, conf, box = d.cls, d.conf, d.xyxy

    return cls, conf, box

# Define YOLOv8 detection function
def yolov8_detection(model, image):
    results = model.predict(image, imgsz=640, conf=0.5, verbose=False)  # Predict with YOLOv8
    result = results[0].cpu()  # Move results to CPU
    box = result.boxes.xyxy.numpy()  # Extract bounding boxes
    conf = result.boxes.conf.numpy()  # Extract confidence scores
    cls = result.boxes.cls.numpy().astype(int)  # Extract class IDs as integers

    return cls, conf, box

# Function to draw boxes on the image
def draw_box(image, detection_output, label_map, colors):
    for cls, conf, box in detection_output:
        color = colors[cls]
        x1, y1, x2, y2 = map(int, box)
        label = f"{label_map[cls]} {conf:.2f}"
        cv2.rectangle(image, (x1, y1), (x2, y2), color, 2)
        cv2.putText(image, label, (x1, y1-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)
    return image

# Function to draw FPS on the image
def draw_fps(fps, image):
    cv2.putText(image, f"FPS: {fps:.2f}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)
    return image

# Function to generate output video name
def get_name(source):
    return pathlib.Path(source).stem

# Function to generate output save path
def get_save_path(file_name, name):
    return f"{name}/{file_name}_output_yolov5n.mp4"

# Define the main detection function
def detection(model_path, source, name, max_frames=500):
    # Check if the model is a TensorRT engine or YOLO model
    file_extension = pathlib.Path(model_path).suffix

    if file_extension == ".engine":
        model = AutoBackend(model_path, device=torch.device('cuda:0'), fp16=True)
        model.warmup()  # Warmup TensorRT model
    else:
        model = YOLO(model_path)  # Load YOLO model

    # Get class names and generate random colors for each class
    label_map = model.names
    COLORS = [[random.randint(0, 255) for _ in range(3)] for _ in label_map]

    # Set up video capture
    video_cap = cv2.VideoCapture(source)
    total_frames = min(int(video_cap.get(cv2.CAP_PROP_FRAME_COUNT)), max_frames)  # Limit to max_frames
    frame_width = int(video_cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(video_cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

    # Prepare for frame-by-frame processing
    frame_count = 0
    total_fps = 0
    video_frames = []
    fps_values = []
    detection_counts = []

    # Loop through each frame in the video
    while video_cap.isOpened() and frame_count < max_frames:
        ret, frame = video_cap.read()
        if not ret:
            break

        start = time.time()  # Start timing

        # Perform detection based on model type
        if file_extension == ".engine":
            cls, conf, box = tensorrt_detection(model, source, frame)
        else:
            cls, conf, box = yolov8_detection(model, frame)

        # Pack detections and annotate the frame
        detection_output = list(zip(cls, conf, box))
        image_output = draw_box(frame, detection_output, label_map, COLORS)

        end = time.time()  # End timing
        fps = 1 / (end - start)  # Calculate FPS
        frame_count += 1
        total_fps += fps

        # Annotate FPS on the frame
        image_output = draw_fps(total_fps / frame_count, image_output)
        video_frames.append(image_output)  # Save annotated frame

        # Collect FPS and detection counts for graphing
        fps_values.append(fps)
        detection_counts.append(len(cls))

        print(f"Processed frame {frame_count}/{total_frames}")  # Log progress

    avg_fps = total_fps / frame_count  # Calculate average FPS
    print(f"Average FPS: {avg_fps}")

    # Save the annotated video
    file_name = get_name(source)
    save_path = get_save_path(file_name, name)
    out = cv2.VideoWriter(save_path, cv2.VideoWriter_fourcc(*'XVID'), int(avg_fps), (frame_width, frame_height))

    for frame in video_frames:
        out.write(frame)

    out.release()  # Release the writer
    print(f"Video saved at: {save_path}")

    # Plot FPS graph (Line Graph)
    plt.figure(figsize=(10, 5))
    plt.plot(fps_values, label='FPS', color='blue')
    plt.title("FPS Over Time")
    plt.xlabel("Frame Number")
    plt.ylabel("FPS")
    plt.legend()
    plt.grid(True)
    plt.show()

    # Plot number of detections graph (Line Graph)
    plt.figure(figsize=(10, 5))
    plt.plot(detection_counts, label='Detections', color='green')
    plt.title("Number of Detections Over Time")
    plt.xlabel("Frame Number")
    plt.ylabel("Number of Detections")
    plt.legend()
    plt.grid(True)
    plt.show()

# Paths for your model and video
model_path = '/content/last.pt'  # Path to your YOLO model
video_path = '/content/drive/MyDrive/video1.mp4'    # Path to your video
output_folder = "/content"  # Folder for saving output

# Run detection for first 500 frames
detection(model_path, video_path, output_folder, max_frames=500)
```
